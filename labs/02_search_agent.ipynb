# Fake Search Agent
Tiny demo using a custom search tool.

```python
from langchain_openai import ChatOpenAI
from langchain.agents import Tool, AgentExecutor, create_react_agent

def fake_search(query):
    return f'Results for {query}: A, B, C'

search_tool = Tool(name='search', func=fake_search, description='Fake search tool')
llm = ChatOpenAI(model='gpt-4o-mini')
agent = create_react_agent(llm, [search_tool], 'Use the search tool when needed.')
executor = AgentExecutor(agent=agent, tools=[search_tool])
executor.invoke({'input':'Search for AI agents'})
```
